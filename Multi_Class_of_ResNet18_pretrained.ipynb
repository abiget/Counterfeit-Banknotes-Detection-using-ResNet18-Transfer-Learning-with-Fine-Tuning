```python
import os
import torch
import pandas as pd
import numpy as np
# from torchvision.io import read_image
from PIL import Image
from torch.utils.data import Dataset
from torchvision import datasets
from torchvision.transforms import ToTensor
import matplotlib.pyplot as plt


class BNotesImageDataset(Dataset):
    def __init__(self, annotations_file=None, img_dir=None, transform=None, target_transform=None):
        self.img_labels = pd.read_csv(annotations_file, usecols=['file_names', 'target_class'])
        self.img_dir = img_dir
        self.transform = transform
        self.target_transform = target_transform
    def __len__(self):
        return len(self.img_labels)

    def __getitem__(self, idx):
        img_path = os.path.join(self.img_dir, self.img_labels.iloc[idx, 0])
        # image = read_image(img_path)
        image = Image.open(img_path)
        label = self.img_labels.iloc[idx, 1]
        if self.transform:
            image = self.transform(image)
        if self.target_transform:
            label = self.target_transform(label)
        return image, label
```

```python
import torchvision
from torchvision import models, transforms
import torchvision.datasets as dset
# import dataloader_script 
# from config_all import Config
import torch

# Now
torch.manual_seed(17)#3407
def get_train_dataloader_with_transform():
  #training transform(augmentation)
  train_data_transform = transforms.Compose([
      transforms.RandomHorizontalFlip(),
      transforms.RandomVerticalFlip(),
      transforms.RandomRotation(45),
      transforms.Resize((224,224)),
      transforms.GaussianBlur(kernel_size=(5, 9), sigma=(0.1, 5)),
      transforms.ColorJitter(brightness=.5, contrast=.5, saturation=.5),
      transforms.ToTensor(),
      # AddGaussianNoise(), This make the image barely visible
      transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225]), #roughly each pixel is with in [-2, 2]
  ])

  #----------------------------------------------------------------
  # training data
  # train_folder_dataset = dset.ImageFolder(root=Config.train_dir)
  # Load the the dataset from raw image folders
  train_dataset = BNotesImageDataset(Config.train_csv,  img_dir=Config.train_dir,
                                                      transform=train_data_transform
                                                      )
  train_loader = torch.utils.data.DataLoader(train_dataset, 
                                                  batch_size=Config.batch_size,
                                                  shuffle=True, 
                                                  num_workers=2)
  return train_loader       

def get_test_dataloader():
  #validation transform test data
  val_data_transform = transforms.Compose([
      transforms.Resize((224,224)),
      transforms.ToTensor(),
      transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])
  ])

  # validation data
  # val_folder_dataset = dset.ImageFolder(root=Config.val_dir) 
  test_dataset = BNotesImageDataset(Config.test_csv, img_dir=Config.test_dir,
                                                    transform=val_data_transform,
                                                    )
  test_loader = torch.utils.data.DataLoader(test_dataset, 
                                                  batch_size=Config.batch_size,
                                                  shuffle=False, 
                                                  num_workers=2)

  return test_loader                                        

def get_val_dataloader():
  #validation transform validation data
  validation_data_transform = transforms.Compose([
      transforms.Resize((224,224)),
      # transforms.GaussianBlur(kernel_size=(5, 9), sigma=(0.1, 5)),
      # transforms.ColorJitter(brightness=.5, contrast=.5, saturation=.5, hue=.005),
      transforms.ToTensor(),
      transforms.Normalize([0.485, 0.456, 0.406], [0.229, 0.224, 0.225])                                               
  ])

  validation_dataset = BNotesImageDataset(Config.val_csv, img_dir=Config.val_dir,
                                                          transform=validation_data_transform)
  validation_loader = torch.utils.data.DataLoader(validation_dataset, 
                                                  batch_size=Config.batch_size,
                                                  shuffle=True, 
                                                  num_workers=2)
  return validation_loader
```

```python
#training the model start here -----------------

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
#loss function
# loss_function = nn.CrossEntropyLoss()

# train_epoch_number = []
train_loss_history = [] 
train_accuracy_history = []

def train_model(model, loss_function, optimizer, data_loader, epoch):
    # set model to training mode
    model.train()

    current_loss = 0.0
    current_acc = 0

    # iterate over the training data
    for i, (inputs, labels) in enumerate(data_loader):
        # send the input/labels to the GPU
        inputs = inputs.to(device)
        labels = labels.to(device)
        # print('labele:', labels)
        # zero the parameter gradients
        optimizer.zero_grad()

        with torch.set_grad_enabled(True):
            # forward
            outputs = model(inputs)
            _, predictions = torch.max(outputs, 1)
            loss = loss_function(outputs, labels)
            # loss = loss_function(outputs, labels)
            # # predictions_softmax = torch.log_softmax(outputs, dim = 1)
            # _, predictions = torch.max(outputs, dim= 1 )

            # backward
            loss.backward()
            optimizer.step()
        # print('Prediction', predictions)
        # statistics
        current_loss += loss.item() * inputs.size(0)
        current_acc += torch.sum(predictions == labels.data)

    total_loss = current_loss / len(data_loader.dataset)
    total_acc = current_acc.double() / len(data_loader.dataset)
    #=================================================================
    train_loss_history.append(total_loss)
    train_accuracy_history.append((total_acc.cpu()*100).item())
    #=================================================================

    print('Ava:Train Loss: {:.4f}; Ava:Accuracy: {:.4f}'.format(total_loss, total_acc*100))
```

```python
def calculate_tpr_fpr(y_real, y_pred):
    '''
    Calculates the True Positive Rate (tpr) and the True Negative Rate (fpr) based on real and predicted observations
    
    Args:
        y_real: The list or series with the real classes
        y_pred: The list or series with the predicted classes
        
    Returns:
        tpr: The True Positive Rate of the classifier
        fpr: The False Positive Rate of the classifier
    '''
    
    # Calculates the confusion matrix and recover each element
    cm = confusion_matrix(y_real, y_pred)
    TN = cm[0, 0]
    FP = cm[0, 1]
    FN = cm[1, 0]
    TP = cm[1, 1]
    
    # Calculates tpr and fpr
    tpr =  TP/(TP + FN) # sensitivity - true positive rate
    fpr = 1 - TN/(TN+FP) # 1-specificity - false positive rate
    
    return tpr, fpr
```

```python
def get_all_roc_coordinates(y_real, y_proba):
    '''
    Calculates all the ROC Curve coordinates (tpr and fpr) by considering each point as a treshold for the predicion of the class.
    
    Args:
        y_real: The list or series with the real classes.
        y_proba: The array with the probabilities for each class, obtained by using the `.predict_proba()` method.
        
    Returns:
        tpr_list: The list of TPRs representing each threshold.
        fpr_list: The list of FPRs representing each threshold.
    '''
    tpr_list = [0]
    fpr_list = [0]
    for i in range(len(y_proba)):
        threshold = y_proba[i]
        y_pred = y_proba >= threshold
        tpr, fpr = calculate_tpr_fpr(y_real, y_pred)
        tpr_list.append(tpr)
        fpr_list.append(fpr)
    return tpr_list, fpr_list
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
def plot_roc_curve(tpr, fpr, auc_value, ax = None):
    '''
    Plots the ROC Curve by using the list of coordinates (tpr and fpr).
    
    Args:
        tpr: The list of TPRs representing each coordinate.
        fpr: The list of FPRs representing each coordinate.
        scatter: When True, the points used on the calculation will be plotted with the line (default = True).
    '''
    if ax == None:
        plt.figure(figsize = (10, 10))
        ax = plt.axes()

    # plt.legend(True)
    sns.lineplot(x = fpr, y = tpr, linestyle='--', color='orange', label="AUC="+str(auc_value), ax = ax)
    sns.lineplot(x = [0, 1], y = [0, 1], linestyle='--', color = '#8B8000', ax = ax)
    plt.xlabel("False Positive Rate")
    plt.ylabel("True Positive Rate")
```
